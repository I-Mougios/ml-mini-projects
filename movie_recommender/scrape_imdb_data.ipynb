```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import datetime


# Set a counter for delay that count every request(click)
counter = 0

# Set a today value
today = datetime.date.today().strftime("%d/%m/%Y")

# Set the df
df = pd.DataFrame(
    columns=[
        "movieId",
        "imdb_rating",
        "imdb_raters",
        "imdb_popularity_step",
        "pop_status(g=green(up),r=red(down))",
        "imdb_popularity_current",
        "imdb_directors",
        "imdb_writers",
        "imdb_stars",
        "user_reviews",
        "critic_reviews",
    ]
)

# create a new Chrome driver
driver = webdriver.Chrome()
driver.maximize_window()

wait = WebDriverWait(driver, 10)

movieid_df = pd.read_csv(
    r"C:\Users\johnx\OneDrive\Desktop\JohnXasa\Masters\Projects\Data Analytics and ML\links.csv",
    dtype={"tmdbId": str, "imdbId": str},
)


for index_count in range(len(movieid_df)):
    try:
        url = "http://www.imdb.com/title/tt" + movieid_df["imdbId"][index_count]
        # navigate to the login page
        driver.get(url)
        if index_count == 0:
            wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="__next"]/div/div/div[2]/div/button[2]')))
            cokkies_btn = driver.find_element(By.XPATH, '//*[@id="__next"]/div/div/div[2]/div/button[2]')
            cokkies_btn.click()

        try:
            wait.until(
                EC.presence_of_element_located(
                    (
                        By.XPATH,
                        '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[2]/div[2]/div/div[1]/a/span/div/div[2]',
                    )
                )
            )
            imdb_rating_data = driver.find_element(
                By.XPATH,
                '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[2]/div[2]/div/div[1]/a/span/div/div[2]',
            ).text
            imdb_rating = imdb_rating_data.split("\n")[0]
            imdb_raters = imdb_rating_data.split("\n")[-1]
        except:
            wait.until(
                EC.presence_of_element_located(
                    (
                        By.XPATH,
                        '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div/div[1]/a/span/div/div[2]',
                    )
                )
            )
            imdb_rating_data = driver.find_element(
                By.XPATH,
                '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div/div[1]/a/span/div/div[2]',
            ).text
            imdb_rating = imdb_rating_data.split("\n")[0]
            imdb_raters = imdb_rating_data.split("\n")[-1]
        # print(imdb_rating)
        # print(imdb_raters)

        try:
            imdb_popularity = driver.find_element(
                By.XPATH,
                '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[2]/div[2]/div/div[3]/a/span/div/div[2]',
            ).text
            imdb_popularity_step = imdb_popularity.split("\n")[1]
            pop_status = driver.find_element(
                By.XPATH,
                '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[2]/div[2]/div/div[3]/a',
            )
            pop_status_svg = pop_status.find_element(By.TAG_NAME, "svg")
            stat = pop_status_svg.get_attribute("class")
            if "up" in stat:
                pop_stat = "g"
            elif "down" in stat:
                pop_stat = "r"
            else:
                pop_stat = "-"

            imdb_popularity_current = imdb_popularity.split("\n")[0]
        except:
            imdb_popularity_step = "-"
            imdb_popularity_current = "-"
            pop_stat = "-"
        # print(imdb_popularity_step)
        # print(pop_stat)
        # print(imdb_popularity_current)

        try:
            imdb_director_data = driver.find_element(
                By.XPATH,
                '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[1]/div',
            )
        except:
            imdb_director_data = driver.find_element(
                By.XPATH,
                '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[4]/div[2]/div[1]/section/div[2]/div/ul/li[1]/div',
            )
        imdb_dir = imdb_director_data.find_elements(By.TAG_NAME, "li")
        imdb_directors = ""
        for dir in imdb_dir:
            dirtemp = dir.find_element(By.TAG_NAME, "a").text
            if imdb_directors == "":
                imdb_directors = dirtemp
            else:
                imdb_directors = imdb_directors + "|" + dirtemp
        # print(imdb_directors)

        try:
            imdb_writer_data = driver.find_element(
                By.XPATH,
                '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[2]/div',
            )
        except:
            imdb_writer_data = driver.find_element(
                By.XPATH,
                '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[4]/div[2]/div[1]/section/div[2]/div/ul/li[2]/div',
            )
        imdb_wri = imdb_writer_data.find_elements(By.TAG_NAME, "li")
        imdb_writers = ""
        for wri in imdb_wri:
            writemp = wri.find_element(By.TAG_NAME, "a").text
            if imdb_writers == "":
                imdb_writers = writemp
            else:
                imdb_writers = imdb_writers + "|" + writemp
        # print(imdb_writers)

        try:
            try:
                imdb_stars_data = driver.find_element(
                    By.XPATH,
                    '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[1]/section/div[2]/div/ul/li[3]/div',
                )
            except:
                imdb_stars_data = driver.find_element(
                    By.XPATH,
                    '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[4]/div[2]/div[1]/section/div[2]/div/ul/li[3]/div',
                )
            imdb_sta = imdb_stars_data.find_elements(By.TAG_NAME, "li")
            imdb_stars = ""
            for sta in imdb_sta:
                statemp = sta.find_element(By.TAG_NAME, "a").text
                if imdb_stars == "":
                    imdb_stars = statemp
                else:
                    imdb_stars = imdb_stars + "|" + statemp
        except:
            imdb_stars = imdb_writers
            imdb_writers = "-"
        # print(imdb_stars)

        try:
            reviewers_data = driver.find_element(
                By.XPATH, '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[3]/div[2]/div[2]/ul'
            )
            reviewers = reviewers_data.find_elements(By.TAG_NAME, "li")
            try:
                user_reviews = reviewers[0].text.split("\n")[0]
            except:
                user_reviews = reviewers[0].text
            try:
                critic_reviews = reviewers[1].text.split("\n")[0]
            except:
                critic_reviews = reviewers[1].text
        except:
            user_reviews = "-"
            critic_reviews = "-"
        # print(user_reviews)
        # print(critic_reviews)

        new_row = [
            movieid_df["movieId"][index_count],
            imdb_rating,
            imdb_raters,
            imdb_popularity_step,
            pop_stat,
            imdb_popularity_current,
            imdb_directors,
            imdb_writers,
            imdb_stars,
            user_reviews,
            critic_reviews,
        ]
        new_df = pd.DataFrame([new_row], columns=df.columns)
        df = pd.concat([df, new_df], ignore_index=True)
    except:
        new_row = [movieid_df["movieId"][index_count], "-", "-", "-", "-", "-", "-", "-", "-", "-", "-"]
        new_df = pd.DataFrame([new_row], columns=df.columns)
        df = pd.concat([df, new_df], ignore_index=True)
        continue

df.to_csv(r"C:\Users\johnx\OneDrive\Desktop\JohnXasa\Masters\Projects\Data Analytics and ML\imdb_data.csv")
driver.close()
```
